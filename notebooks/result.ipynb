```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
def infer_heuristic(descriptor: str) -> str:
    if "lexical" in descriptor:
        return "Lexical Overlap"
    if "subsequence" in descriptor:
        return "Subsequence"
    if "constituent" in descriptor:
        return "Constituent"


def infer_split(descriptor: str) -> str:
    if "train" in descriptor:
        return "Train"
    else:
        return "Eval"


def infer_quantity(descriptor: str) -> str:
    if "accuracy" in descriptor:
        return "Accuracy"
    else:
        return "Loss"


def infer_entailment(descriptor: str) -> str:
    if "non-entailment" in descriptor:
        return "No-Entailment"
    else:
        return "Entailment"
```

```python
rows = []
for entry in data:
    eraser = entry.pop("eraser")
    for key, value in entry.items():
        if infer_quantity(key) == "Accuracy":
            rows.append(
                {
                    "Eraser": eraser,
                    "Split": f"{infer_heuristic(key)}-{infer_split(key)}-{infer_entailment(key)}",
                    "Accuracy": value,
                }
            )
    entry["eraser"] = eraser
df = pd.DataFrame(rows)
df.head()
```
